```python
import numpy as np
import argparse
import pandas as pd
import glob
import datetime
import matplotlib.pyplot as plt
import act
import os
import sys
sys.path.append('../')
from tempfile import TemporaryDirectory
import altair as alt
```

```python
USERNAME = os.getenv("ARM_USERNAME")
TOKEN = os.getenv("ARM_TOKEN")
SAIL_DATA_STREAM = 'gucdlprofwind4newsM1.c1'
DATA_STREAM_FILEEXT = '.nc'
date = (datetime.datetime.today() - datetime.timedelta(days=30)).strftime('%Y-%m-%d')
```

```python
startdate = date
enddate = date

with TemporaryDirectory() as temp_dir:
    act.discovery.download_data(USERNAME, TOKEN, SAIL_DATA_STREAM, startdate, enddate, output=temp_dir)
    dl_prof_files = glob.glob(''.join([temp_dir, '/', SAIL_DATA_STREAM, '*'+DATA_STREAM_FILEEXT]))
    print(len(dl_prof_files))
    dl_prof = act.io.armfiles.read_netcdf(dl_prof_files)
```

```python
src_prof = dl_prof.to_dataframe().reset_index()
```

```python
src_prof = src_prof[src_prof['time'].dt.minute < 5]
```

```python
src_prof['time'] = src_prof['time'].dt.strftime('%Y-%m-%d - %H')
```

```python
src = src_prof.head(500)[['time', 'height','met_wspd']]
```

```python
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
```

```python
src = src.query('time == "2023-01-08 - 00"')
```
