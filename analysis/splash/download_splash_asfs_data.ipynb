```python
import ftplib
import os
import urllib
```

```python
# download dir
avp_download_dir = "/data2/elilouis/sublimationofsnow/asfs/avp"
kps_download_dir = "/data2/elilouis/sublimationofsnow/asfs/kps"

ftp_url = 'ftp1.esrl.noaa.gov'
# Avery Picnic product
avp_url_level2 = f'Observations/Campaigns/SPLASH/asfs50/2_level_product/'
# Kettle Ponds product
kps_url_level2 = f'Observations/Campaigns/SPLASH/asfs30/2_level_product/'
# Avery Picnic product
avp_url_level1 = f'Observations/Campaigns/SPLASH/asfs50/1_level_ingest/'
# Kettle Ponds product
kps_url_level1 = f'Observations/Campaigns/SPLASH/asfs30/1_level_ingest/'
```

```python
ftp = ftplib.FTP(ftp_url)
ftp.login()
```

# Download Level 2 Products

```python
kps_file_ls = ftp.nlst("Observations/Campaigns/SPLASH/asfs30/2_level_product")
kps_file_ls = [f for f in kps_file_ls if 'sledseb' in f]

for i, f in enumerate(kps_file_ls):
    print(f"Downloading KPS file {i+1} of {len(kps_file_ls)}")
    filename = f.split("/")[-1]
    remote_location = os.path.join('ftp://', ftp_url, kps_url_level2, filename)
    local_location = os.path.join(kps_download_dir, filename)
    # print(remote_location)
    # print(local_location)
    urllib.request.urlretrieve(remote_location, local_location)

avp_file_ls = ftp.nlst("Observations/Campaigns/SPLASH/asfs50/2_level_product")
avp_file_ls = [f for f in avp_file_ls if 'sledseb' in f]

for i, f in enumerate(avp_file_ls):
    print(f"Downloading AVP file {i+1} of {len(avp_file_ls)}")
    filename = f.split("/")[-1]
    remote_location = os.path.join('ftp://', ftp_url, avp_url_level2, filename)
    local_location = os.path.join(avp_download_dir, filename)
    # print(remote_location)
    # print(local_location)
    urllib.request.urlretrieve(remote_location, local_location)
```
