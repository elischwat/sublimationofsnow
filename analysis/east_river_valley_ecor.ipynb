```python
import os
import act
import glob
```

```python
ecor_gothic = 'guc30ecorM1.b1'
ecor_kp = 'guc30ecorS3.b1'

met_gothic = 'gucmetM1.b1'

start_date = '2022-10-31'
end_date = '2022-11-04'
```

```python
username = os.getenv("ARM_USERNAME")
token = os.getenv("ARM_TOKEN")
username,token
```

```python
act.discovery.download_data(username, token, ecor_gothic, start_date, end_date)
act.discovery.download_data(username, token, ecor_kp, start_date, end_date)

act.discovery.download_data(username, token, met_gothic, start_date, end_date)
```

```python
ecor_gothic_ds = act.io.armfiles.read_netcdf(
    glob.glob(os.path.join(ecor_gothic, '*.cdf'))
)
```

```python
ecor_kps_ds = act.io.armfiles.read_netcdf(
    glob.glob(os.path.join(ecor_kp, '*.cdf'))
)
```

```python
met_gothic_ds = act.io.armfiles.read_netcdf(
    glob.glob(os.path.join(met_gothic, '*.cdf'))
)
```
