```python
# flux_df = flux_df[(flux_df['date'].dt.year == 2022) & (flux_df['date'].dt.month == 2) & (flux_df['date'].dt.day == 14)]
flux_df = flux_df.query("height == 3")
```

in each day:
* sum up all moisture fluxes
* divide by latent heat of sublimation

```python
flux_df['qc_LE'].unique()
```

```python
latent_heat_of_sublimation_ice = 2838000 #J/kg

weight_cubic_meter_of_water = 1000 # kg/m^3
```

```python
daily_mean_flux_df = flux_df.groupby(flux_df['date'].dt.date).mean() # latent heat flux in joules/m^2/sec
daily_mean_flux_df['sublimation_rate'] = daily_mean_flux_df['LE'] / latent_heat_of_sublimation_ice # kg/m^2/sec
daily_mean_flux_df['sublimation_rate'] = daily_mean_flux_df['sublimation_rate'] / weight_cubic_meter_of_water # m/sec

daily_mean_flux_df['daily_cumulative_sublimation'] = daily_mean_flux_df['sublimation_rate'] * 1000 * 86400 # mm/day
daily_mean_flux_df = daily_mean_flux_df.reset_index()
```
